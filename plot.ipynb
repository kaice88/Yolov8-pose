```python
import matplotlib.pyplot as plt

# Danh sách accuracy của tập huấn luyện theo epoch
pose_loss = df["          val/pose_loss"].values

# Số epoch tương ứng với danh sách accuracy
epochs = range(1, len(pose_loss) + 1)

# Vẽ biểu đồ
plt.figure(figsize=(10, 6))
plt.plot(epochs, pose_loss, 'b')
plt.title('Validate pose loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```
